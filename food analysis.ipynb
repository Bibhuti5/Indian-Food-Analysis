```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Import data set 

```python
df=pd.read_csv("indian_food.csv")
df.head()
```

### Finding any Null Value

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

### Replaceing -1 to nan

```python
df=df.replace(-1,np.nan)
df=df.replace('-1',np.nan)
```

```python
df.isnull().sum()
```

```python
df.shape
```

## Ploting State Wise Ingredients Use

```python
plt.figure(figsize = (15,15))
sns.countplot(y='state', data=df,palette=sns.color_palette())
plt.title('State wise Ingredients', fontsize = 50, fontweight = 'bold', color='Black')
plt.xlabel('Ingredients total no of use', size = 50, color='green')
plt.ylabel('All State name', size =50, color='blue')
plt.xticks(size =20, color='black')
plt.yticks(size =20, color='black')
plt.xlim([0,40])
plt.show()
```

## State wise Diet

```python
plt.figure(figsize=(15,15))
sns.countplot('diet', data = df, palette=sns.color_palette())
plt.title("State wise Diet",fontsize=30, fontweight = 'bold', color='black')
plt.ylabel("Percentage", size = 25, color='black')
plt.xlabel("Diet", size = 25, color='black')
plt.xticks(size = 20, color='black')
plt.yticks(size = 20, color='black')
plt.show()
```
